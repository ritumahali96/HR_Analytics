# DATA CLEANING , JUSTIFICATION AND FEATURE ENGINEERING

```python
#Let us see at the percent of null value missing 
percent_missing = df.isnull().sum() * 100 / len(df)
percent_missing
```

Total percent missing = percent_missing(Base_pay + openingbalance + Total_Sales) = (0.46+29.52+0.32)% = 30.3 %

In the case of a high number of outliers in a dataset, it is recommended to use the median to replace the nan values. Since we have 30.3% percent of missing values. We will use median in the next step.

```python
# Replacing the missing values of the columns; Base_pay, openingbalance, Total_Sales
df['Total_Sales']=df['Total_Sales'].replace(np.nan,df['Total_Sales'].median())
df['openingbalance']=df['openingbalance'].replace(np.nan,df['openingbalance'].median())
df['Base_pay']=df['Base_pay'].replace(np.nan,df['Base_pay'].median())
```

From the above, we can see that we have outliers in Base_pay and openingbalance. Also, we have more than 300 outliers in openingbalance. So, we need to keep the column for furthur interpretations and can't be removed.

```python
# Treating Outliers of Base_pay using IQR(Inter Quartile Range) method
q1 = df.Base_pay.quantile(0.25) # Calculate q1
q3 = df.Base_pay.quantile(0.75) # Calculate q2
print("q1: ",q1)
print("q3: ",q3)

IQR = q3-q1       # Calculate q3
print("IQR: ",IQR)
#Finding the lower and upper limits as Q1 – 1.5 IQR and Q3 + 1.5 IQR
low_lim = q1-(1.5*IQR)
up_lim = q3+(1.5*IQR)
print("lower limit: ",low_lim)

print("upper limit: ",up_lim)
```

```python
# The datapoints that are left after treating outliers
df = df[(df.Base_pay>low_lim) & (df.Base_pay<up_lim)]
df.shape
```

# MODEL BUILDING

# a)Trying Linear Regression Algorithm:-

Reason of choosing linear regression algorithm is: 
We are finding out the effect of Input variables on Target variable.
Finding out the change in Target variable with respect to one or more input variable is the best case
for using Linear Regression Algorithm.

```python
from sklearn import datasets, linear_model, metrics

# We need to find how salary is varying with respect to other considerable factors. 
# So, Salary is dependent variable. The remaining columns are considered to be independent here.

X = new_data.drop(columns=['Salary'])
y = new_data['Salary']
```

```python
# splitting X and y into training and testing sets
'''We can create both the training and testings sets in a one-liner by passing to train_test_split() 
the modelling DataFrame along with the fraction that should be included in the 
testing set.  
We set a random_state so that the results are reproducible, that is every time we run the code, 
the same instances will be included in the training and testing sets 
respectively. 
The method returns a tuple with two DataFrames containing the training and testing examples.'''

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=0)

#X = Input variables

#y = Target variable

#test_size = The ratio of the dataset that’ll be used in test set. If there are 100 rows, 70 rows will 
#be used in train set and 30 in test set.'''

#random_state = It is provided just for the sake of obtaining the same result every time we run the 
#code.
#It ensures that the rows in train and test set remain same each time.'''
```

```python
from sklearn.linear_model import LinearRegression # Importing the model from sklearn
linear_regressor = LinearRegression() # create object for the class / initialising & storing the model in a variable
```

```python
#fit model to our  training data i.e. learn coefficients
linear_regressor.fit(X_train , y_train) # Training the model using fit method
y_pred = linear_regressor.predict(X_test)
y_pred
```

```python
# importing r2_score module
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
# predicting the accuracy score
score=r2_score(y_test,y_pred)
print("r2 score is ",score)
print("mean_sqrd_error is ",mean_squared_error(y_test,y_pred))
print("root_mean_squared error of is ",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
linear_regressor.score(X_test, y_test) # Re-checking the accuracy score
```

# c)TRYING RANDOM FOREST TECHNIQUE FOR THE MODEL:-

```python
from sklearn.ensemble import RandomForestRegressor

X = new_data.drop(columns=['Salary']) # Input Variable
y = new_data['Salary'] # Target varible

# splitting X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=1)

# create linear regression object

rf1=RandomForestRegressor(n_estimators=500,random_state=1)

# train the model using the training sets
rf1.fit(X_train, y_train)
#Predict the target
y_pred =  rf1.predict(X_test)
y_pred
```

```python
# importing r2_score module
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
# predicting the accuracy score
score=r2_score(y_test,y_pred)
print("r2 score is ",score)
print("mean_sqrd_error is ",mean_squared_error(y_test,y_pred))
print("root_mean_squared error of is ",np.sqrt(mean_squared_error(y_test,y_pred)))
```
